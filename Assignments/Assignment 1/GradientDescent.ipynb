```python
import pandas as pd 
import matplotlib.pyplot as plt

def gradient_descent(m_current=0, b_current=0, epochs=5000, learning_rate=0.01):
    data = pd.read_csv('4-data.txt', sep='\s+')
    X = data.Age
    y = data.Distance
    X /= 10
    y /=100
    N = float(len(y))
    for i in range(epochs):
        y_current = (m_current * X) + b_current
        cost = sum([data**2 for data in (y - y_current)]) / N
        m_gradient = -(2/N) * sum(X * (y - y_current))
        b_gradient = -(2/N) * sum(y - y_current)
        m_current = m_current - (learning_rate * m_gradient)
        b_current = b_current - (learning_rate * b_gradient)
        
    plt.title('Gradient Descent \n Predicted:y='+str(m_current)[0:6]+'x +'+str(b_current)[0:6] );
    plt.xlabel('Age')
    plt.ylabel('Distance')
    plt.grid(True) # 显示网格
    plt.plot(X,y,'r.') # 绘图
    plt.plot(X,y_current,'g-') # 绘图
    plt.show() # 显示图像

gradient_descent()
```
